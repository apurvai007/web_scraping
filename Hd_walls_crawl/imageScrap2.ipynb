```python
import requests as req
from bs4 import BeautifulSoup
import ssl
import os
```

```python
# Ignore ssl certificate error
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input("Enter url : ")
r1  = req.get(url)
soup  = BeautifulSoup(r1.text,"html.parser")


link = []

x = soup.select('img[src^="https://images.pexels.com/photos"]')

for img in x:
    link.append(img['src'])
#print(link)

os.mkdir("hd_walls")

count = 0
for i,url in enumerate(link):
    
    img_data = req.get(url).content
    with open("hd_walls/"+str(i)+".jpeg","wb+") as f:
        f.write(img_data)
        count += 1
    
    if( count > 20):
        f.close()
        
                              
print("done -- go check hd_walls folder")                            
                              
                              
                              
                              
```
