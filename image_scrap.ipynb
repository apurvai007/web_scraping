```python
from bs4 import *
import requests as rq
import os
import ssl

# Ignore ssl certificate error
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE


r1 = rq.get("https://www.pexels.com/@hiteshchoudhary")
soup1 = BeautifulSoup(r1.text,"html.parser")


links = []
#https://images.pexels.com/photos/
# 2756843/pexels-photo-2756843.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500
x = soup1.select('img[src^="https://images.pexels.com/photos"]')


for img in x:
    #regex to get url of images
    links.append(img['src'])

# for link in links:
#     print(link)
"""VERIFIED LINKS IN LIST"""

os.mkdir("scraped_pics")
i = 1

for index,link in enumerate(links):
    if i<=10:
        img_data = rq.get(link).content
        with open("scraped_pics//"+str(index+1)+".jpeg","wb+") as f:
            f.write(img_data)
        i +=1
    else:
        f.close()
        break
            


```
